```python
import cv2
import pytesseract
import re
import os
import json

images = os.listdir('images')

list_of_data = []

for i, image_file in enumerate(images):

    # image_file = 'images/20230104_115403AMByGPSMapCamera.jpg'

    print(f"{i+1}/{len(images)}", image_file)

    # Extract text from image
    img = cv2.imread('images/'+image_file)

    pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'
    config = ('-l eng --oem 1 --psm 3')
    text = pytesseract.image_to_string(img, config=config)

    lat = re.findall(r'(?<=Lat )\d+.\d+', text)
    long = re.findall(r'(?<=Long )\d+.\d+', text)

    print(lat, long)
    list_of_data.append({'image': image_file, 'lat': lat, 'long': long})

with open('data.json', 'w') as outfile:
    json.dump(list_of_data, outfile)

print('All files Done')
```
